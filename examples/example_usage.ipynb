```python
%%capture
import sys

# Añade el directorio principal al path de búsqueda para importar módulos desde esa ubicación
sys.path.insert(0, "..")

from io import BytesIO
import matplotlib.pyplot as plt
import numpy as np

from image_describer import ImageDescriber

plt.rcParams["font.size"] = 10
plt.style.use("seaborn-v0_8")
```

```python
describer = ImageDescriber()
# Try with an online image URL
description1 = describer.describe_image(
    "https://qianwen-res.oss-cn-beijing.aliyuncs.com/Qwen-VL/assets/demo.jpeg"
)
print("Online image description:", description1)
```

```python
# Generate the plot
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)
plt.title("Example Sine Wave")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Save to BytesIO object
img_data = BytesIO()
plt.savefig(img_data, format="png")
img_data.seek(0)

# Call the describe_image method with the BytesIO object
description = describer.describe_image(img_data)
print(description)
plt.show()
```
